```python
import tensorflow as tf
import numpy as np
```

### Set up training data

```python
celsius_q    = np.array([-40, -10,  0,  8, 15, 22,  38,  55,  60,  70],  dtype=np.float32)
fahrenheit_a = np.array([-40,  14, 32, 46, 59, 72, 100, 131, 140, 156],  dtype=np.float32)

for i,c in enumerate(celsius_q):
    print("{:5}°C = {:5}°F".format(c, fahrenheit_a[i]))
```

### Create a Model

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

### Compile the model, with loss and optimizer functions

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.05))
```

### Train the model

```python
history = model.fit(celsius_q, fahrenheit_a, epochs=1000, verbose=False)
```

### Display training stats

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
# Answer: 100×1.8+32=212 
print(model.predict([100.0]))
```
